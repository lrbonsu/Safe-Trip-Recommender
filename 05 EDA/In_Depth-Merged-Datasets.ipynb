```python
#import pandas
import pandas as pd
```

```python
#read in spi file file 
csv1_data = pd.read_csv('new_spi.csv')

# convert into dataframe
spi_df = pd.DataFrame(csv1_data)
```

```python
#read in lpi file file 
csv2_data = pd.read_csv('new_lpi.csv')

# convert into dataframe
lpi_df = pd.DataFrame(csv2_data)
```

```python
#verifying df
spi_df
```

```python
#verifying df
lpi_df
```

```python
#merge both dataframes
merged_df = pd.merge(spi_df, lpi_df, on=['country', 'year'], how='outer', indicator=True)
```

```python
merged_df
```

```python
#identify string to look for
discard = ["both"]
#view df without rows that were merged
not_merged_df = merged_df[~merged_df._merge.str.contains('|'.join(discard))]
```

```python
not_merged_df
```

```python
#show countries that did not merge
not_merged_df.country.unique()
```

```python
#view countries that didn't merge from left dataset
#identify string to look for
right_discard = ["right"]
#view df with left df that didn't merge
left_not_merged = not_merged_df[~not_merged_df._merge.str.contains('|'.join(right_discard))]
```

```python
left_not_merged
```

```python
#show countries that did not merge from left
left_not_merged.country.unique()
```

```python
#view countries that didn't merge from right dataset
#identify string to look for
left_discard = ["left"]
#view df with left df that didn't merge
right_not_merged = not_merged_df[~not_merged_df._merge.str.contains('|'.join(left_discard))]
```

```python
right_not_merged
```

```python
#show countries that did not merge from right
right_not_merged.country.unique()
```
