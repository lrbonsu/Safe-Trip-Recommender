```python
#create new dataframe for lgbt rights, year and country
lgbt_df = df[['area_name', 'year', 'lgbt_rights']]
```

```python
#stack or melt for narrow frame
lgbt_df = pd.melt(lgbt_df, id_vars =['lgbt_rights', 'year'], value_vars='area_name', var_name = 'country_', value_name='COUNTRY')
lgbt_df
#drop unnecessry column
lgbt_df=lgbt_df.drop(columns= ['country_'])
#groupby year and lgbt right score. Get size of scores
lgbt_result=lgbt_df.groupby(['year','lgbt_rights']).size()
lgbt_result_df = pd.DataFrame([lgbt_result])
lgbt_result_df.reset_index()
```

```python
#unstack multi-index and transform series to dataframe
lgbt = lgbt_result_df.unstack()
lgbt2=pd.DataFrame(lgbt).reset_index()
lgbt2
```

```python
#create multi-plot for plotting conditional relationship between year and LGBT rights 
lgbt_graph = sns.FacetGrid(lgbt2, col= "year", col_wrap=3, height=4.5, aspect=1.0)
lgbt_graph.map(sns.barplot, 'lgbt_rights', 0).set(xlabel = 'LGBT Rights', ylabel='count')
```
