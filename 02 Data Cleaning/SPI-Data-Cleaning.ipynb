```python
#import pandas
import pandas as pd
```

```python
#read in csv file
csv_data = pd.read_csv('Cleaned_Team16_Dataset.csv')

# convert csv into dataframe
df = pd.DataFrame(csv_data)
```

```python
#drop rows where country contains the word "World"

#identify partial string to look for
discard = ["World"]

#drop rows that contain the string "index" in the country column #without the tilde, it drops everything except              
df_spi1 = df[~df.country.str.contains('|'.join(discard))]

#reset index of series object and drop original index labels
df_spi2 = df_spi1.reset_index(drop = True)
df_spi2
```

```python
#drop spi_country_code, spi_rank, status, and unique_id columns
df_spi3 = df_spi2.drop(columns=['unique_id', 'status', 'spi_rank', 'spi_country_code'])
#rename spi_year column
df_spi3.rename(columns={'spi_year':'year'}, inplace=True)
df_spi3
```
