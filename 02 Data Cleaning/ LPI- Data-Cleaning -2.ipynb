```python
#import pandas
import pandas as pd
```

```python
#load the xlsx file and read in the 4th sheet
excel_data = pd.read_excel('LPI_Legatum_Prosperity_Index_Indicators_Sheet4.xlsx')

# convert into dataframe
df = pd.DataFrame(excel_data)

#view columns of initial df
df.columns
```

```python
#number of rows and columns in initial df
df.shape
```

```python
#view first 10 rows of initial df
df.head(10)
```

```python
#drop unnecessary columns
df = df.drop(columns = ['pillar_name', 'element_name','rank_2007','rank_2008', 'rank_2009', 'rank_2010','raw_value_2007', 'raw_value_2008', 'raw_value_2009',
       'raw_value_2010', 'raw_value_2011', 'raw_value_2012', 'raw_value_2013',
       'raw_value_2014', 'raw_value_2015', 'raw_value_2016', 'raw_value_2017',
       'raw_value_2018', 'raw_value_2019', 'raw_value_2020', 'raw_value_2021',
        'score_2007', 'score_2008', 'score_2009', 'score_2010', 'rank_2011','rank_2012','rank_2013', 'rank_2014', 
                                        'rank_2015', 'rank_2016', 'rank_2017','rank_2018', 
                                        'rank_2019', 'rank_2020', 'rank_2021'])

#pivot table to make new columns
df_pivot1 = pd.pivot_table (df, values=[ 'score_2011',
                                        'score_2012', 'score_2013', 'score_2014', 'score_2015', 
                                        'score_2016','score_2017', 'score_2018', 'score_2019', 
                                        'score_2020', 'score_2021'],
                            columns= ['area_name','indicator_name']).reset_index()

#view new columns
df_pivot1.columns
```

```python
df_pivot1
```

```python
#stack dataframe to remove multi-index and reset index to set new columns after stack
df_pivot2 = df_pivot1.stack(level = -2).reset_index()

#rename blank column header
df_pivot2.rename(columns={'': 'temp_name'}, inplace=True)
df_pivot2
```

```python
#drop rows where area_name contains the word "index"

#identify partial string to look for
discard = ["index"]

#drop rows that contain the partial string "index" in the area_name column
df_pivot3 = df_pivot2[~df_pivot2.area_name.str.contains('|'.join(discard))]

#reset index of series object and drop original index labels
df_drop_rows = df_pivot3.reset_index(drop = True)
df_drop_rows
```

```python
#create new column to add to dataframe
#create list with values
data_list = ['2011', '2012', '2013','2014', '2015', '2016', '2017', '2018', 
'2019', '2020', '2021']

#convert list into new series
data_series = pd.Series(data_list)

#repeat elements of the series
new_year=data_series.repeat(167)
#reset index of series object and drop original index labels
new_year2 = new_year.reset_index(drop = True)
new_year2
```

```python
#put new column in DataFrame at a specified location. Add values as rows to create new column
#df_pivot3.append(new_year, ignore_index=True)
df_drop_rows.insert(3, "year", new_year2)
```

```python
#show new df with new columns
df_drop_rows
```

```python
#clear index name
df_drop_rows.columns.name = None
#drop unnecessary columns
df_drop_rows = df_drop_rows.drop(columns = ['level_0', 'temp_name'])
df_drop_rows
```

```python
#rename columns that contain apostrophes or quote marks
df_drop_rows.rename(columns={"Women's agency": 'womens_agency', 
                          "Protection of women's workplace, education and family rights": 'protection_womens_rights'}, 
                 inplace=True)
```

```python
#rename remaining columns to keep
df_drop_rows.rename(columns={'Commercial bank branches': 'bank_branches', 'Use of digital payments': 'digital_payments', 
                          'Satisfaction with public transportation': 'satisfaction_with_public_transportation', 
                          'International internet bandwidth':'international_internet_bandwidth', '2G, 3G and 4G network coverage': 'network_coverage', 
                          'Reliability of electricity supply': 'reliable_electricity', 'Reliability of water supply':'reliability_water_supply', 
                          'Generalised interpersonal trust': 'generalised_interpersonal_trust', 'Helped a stranger':'helped_stranger', 'Confidence in local police': 'confidence_in_police', 
                          'Satisfaction with freedom': 'satisfaction_with_freedom','Freedom of opinion and expression': 'freedom_opinion_expression', 
                          'Government media censorship' : 'government_media_censorship', 'Equal treatment and absence of discrimination': 'equal_treatment', 
                          'Non-discriminatory civil justice':'nondiscriminatory_civil_justice', 'LGBT rights' : 'lgbt_rights', 'Freedom of belief and religion': 'freedom_of_belief_religion', 
                          'Government religious intimidation and hostility': 'government_intimidation', 'Two-sided conflict deaths': 'twosided_conflict_deaths', 
                          'Civil and ethnic war': 'civil_ethnic_war', 'Terrorism incidents': 'terrorism_incidents', 'Intentional homicides': 'intentional_homicides', 
                          'Safety walking alone at night': 'safety_walking_alone_night', 'Physical security of women': 'physical_security_women','Property stolen':'property_stolen', 'Respect':'respect'}, 
                 inplace=True)
```

```python
#drop all columns excep for the columns named below
df_drop_rows.drop(df_drop_rows.columns.difference(['area_name','year','bank_branches', 'digital_payments','satisfaction_with_public_transportation', 
                          'international_internet_bandwidth', 'network_coverage','reliable_electricity', 'reliability_water_supply', 
                          'generalised_interpersonal_trust', 'helped_stranger',  'confidence_in_police', 'satisfaction_with_freedom', 'freedom_opinion_expression', 
                          'government_media_censorship', 'equal_treatment','nondiscriminatory_civil_justice', 'lgbt_rights',  'freedom_of_belief_religion', 
                          'government_intimidation', 'twosided_conflict_deaths', 'civil_ethnic_war', 'terrorism_incidents', 'intentional_homicides', 
                          'safety_walking_alone_night', 'physical_security_women','property_stolen', 'womens_agency', 'protection_womens_rights', 'respect']),
                    1, inplace=True)
```

```python
#rename area_name column
df_drop_rows.rename(columns={'area_name':'country'}, inplace=True)
```
